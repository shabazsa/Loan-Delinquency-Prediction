```python
#importing models
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
abc=[]
classifiers=['Logistic Regression','KNN','Decision Tree', 'Random Forest', 'Naive Bayes']
models=[LogisticRegression(),
        KNeighborsClassifier(n_neighbors=21),DecisionTreeClassifier(), 
        RandomForestClassifier(n_estimators=500), GaussianNB()]
for i in models:
    model = i
    model.fit(x, y)
    prediction=model.predict(x1)
    abc.append(prediction)
```

```python
models_dataframe = pd.DataFrame(abc, index=classifiers)
```
