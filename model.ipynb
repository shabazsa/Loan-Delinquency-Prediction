```python
tr = pd.read_csv(root_path + 'dataset/train.csv')
tr.head()
```

```python
# one hot encoding function
```

```python
def one_hot_encoding(dataframe, feature):
    """Accetps a dataframe and returns the dataframe with feature one_hot_encoded"""
    financial_institution_dict = {name: value for value, name in enumerate(set(list(dataframe[feature])))} 
    dataframe[feature] = dataframe[feature].map(financial_institution_dict)
    # tr['financial_institution']
    # financial_institution_dict
    ohe = pd.get_dummies(dataframe[feature])
    dataframe = dataframe.drop(feature, axis=1)
    dataframe = dataframe.join(ohe)
    
    return dataframe
```

```python
data = pd.get_dummies(tr)
```

```python
data = data.drop('loan_id', axis=1)
```

```python
Y = data.m13
```

```python
X = data.drop('m13',axis=1)
X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fit model
melbourne_model.fit(X, Y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```

```python
test_data = pd.read_csv(root_path + 'dataset/test.csv')
test_data = pd.get_dummies(test_data)
test_data = test_data.drop('loan_id', axis=1)
test_data.head()
```

```python
X.head()
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
predictions = melbourne_model.predict(test_data)
```

```python
with open(root_path + 'predictions.csv', 'a') as file:
  file.write('loan_id,m13\n')
  for row,i in enumerate(predictions,1):
    file.write("{},{}\n".format(row,int(i)))
```
